```python
import re
from operator import itemgetter    
 
frequency = {}
open_file = open('d2016.bin', 'r')
file_to_string = open_file.read()

# locate all the individual words in the text string variable
# find all the words that start with a letter (upper-case or lower-case) and followed by a sequence of letters which consist of at least 2 characters and no more than 9 characters.
# \b[A-Za-z][a-z]{2,10}\b
words = re.findall(r'(\b[A-Za-z][a-z]{2,9}\b)', file_to_string)
 
for word in words:
    count = frequency.get(word,0)
    frequency[word] = count + 1
     
# print the key-value pair of the dictionary, showing the word (key) and the number of times it appeared in the list (value):
for key, value in reversed(sorted(frequency.items(), key = itemgetter(1))):
    print (key, value)
```

```python
import pandas as pd 
df = pd.read_csv("zipf.csv")
df
```

```python
from sklearn import preprocessing
import numpy as np# Convert to array
data_array = np.array(df)
print(data_array)
```

```python
#need to normalize each column individually
# first the wiki column

# copy the data
df_max_scaled = df.copy()
  
# apply normalization techniques on Column 1
column = 'wiki'
df_max_scaled[column] = df_max_scaled[column] /df_max_scaled[column].abs().max()
  
# view normalized data
display(df_max_scaled)
```

```python
#need to normalize each column individually
# then the d2016 column

# copy the data
df_max_scaled2 = df_max_scaled.copy()
  
# apply normalization techniques on Column 1
column = 'd2016'
df_max_scaled2[column] = df_max_scaled2[column] /df_max_scaled2[column].abs().max()
  
# view normalized data
display(df_max_scaled2)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# gca stands for 'get current axis'
ax = plt.gca()
df_max_scaled2.plot(kind='line',y='zipf',color='red', ax=ax)
df_max_scaled2.plot(kind='line',y='d2016',color='blue', ax=ax)
df_max_scaled2.plot(kind='line',y='wiki',color='gray', ax=ax)
plt.show()
```

```python
# Conclusion the d2016.data file (blue) does not match the Zipf Distribution (red) as the article claims.
# However, the same process with the Wikipedia article on Zipf Law (gray) does.
```
